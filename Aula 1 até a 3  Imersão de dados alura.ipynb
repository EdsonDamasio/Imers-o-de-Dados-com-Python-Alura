```python
import numpy as np
import pandas as pd

df_salarios = pd.DataFrame({
    'nome': ["Leticia", "Leandro", "Kaique", "Marcos","Everson"],
    'salario': [4000, np.nan, 5000, np.nan,7000]
})

#calcula a média salarial e substitui as nulas pela média e arrendonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(
    df_salarios['salario'].mean())

df_salarios['salario_mediana'] = df_salarios['salario'].fillna(
    df_salarios['salario'].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["segunda", "terça", "quarta", "quinta", "sexta", "sábado", "domingo"],
    "Temperatura": [30, np.nan, np.nan, 28, 27, 30, 28]
})

df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
import pandas as pd
import plotly.express as px

# Assuming df is globally available, or re-read it if the notebook state is reset
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")

#Gráfico Pizza
df_limpo = df.dropna()
df_limpo = df_limpo.assign(work_year = df_limpo['work_year'].astype('int64'))
df_limpo = df_limpo.rename(columns={'experience_level': 'nivel_de_experiencia'})
df_limpo["nivel_de_experiencia"] = df_limpo["nivel_de_experiencia"].replace({
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
})
remoto_contagem = df_limpo['remote_ratio'].value_counts().reset_index()
remoto_contagem.columns = ['remote_ratio', 'count']

fig = px.pie(
    remoto_contagem,
    values='count',
    names='remote_ratio',
    title='Distribuição de Remote Ratio',
    color='remote_ratio'
)

fig.show()
```
