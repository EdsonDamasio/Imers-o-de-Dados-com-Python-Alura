```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
df =df[df['job_title'] == 'Data Scientist']
```

```python
#Níveis de experência
map_experiencia = {
    'EN': 'Junior',
    'MI': 'Pleno',
    'SE': 'Sênior'
}

df['Nível_experiencia'] = df['experience_level'].map(map_experiencia)
df = df.dropna(subset=['Nível_experiencia'])
```

```python
#paises
paises = [
    #América
    'US', 'BR', 'CA', 'MX', 'AR', 'CL', 'CO', 'PE',

    #Europa
    'PT', 'ES', 'IT', 'FR', 'DE', 'GB', 'NL', 'IE',

    #Ásia e Oceania
    'JP', 'TH', 'IN', 'AU'
]

df = df[df['employee_residence'].isin(paises)]
```

```python
#calcular Sálario Médio Anual (USD)
media_salarios_anuais = (
    df.groupby(
        ['employee_residence', 'Nível_experiencia'],
        as_index=False
        ) ['salary_in_usd']
    .mean()
)
```

```python
#Cores padronizadas por nível de experiência
cores_experiencia = {
    'Júnior': 'blue',
    'Pleno': 'green',
    'Sênior': 'red',
}

#Gráfico interativo

fig = px.bar(
    media_salarios_anuais,
    x='employee_residence',
    y='salary_in_usd',
    color='Nível_experiencia',
    barmode='group',
    title='Salário medio Anual Cientista de dados por país e Nível de experiência',
    labels={'employee_residence': 'País', 'salary_in_usd': 'Salário Médio Anual (USD)', 'nivel_experincia': 'Nível de Experiência'
    },
    color_discrete_map=cores_experiencia,
    text_auto='.2s'
)
```

```python
fig.update_layout(
    template='plotly_dark',
    legend_title='Senioridade',
    xaxis_title='País',
    yaxis_title='Salário Médio Anual (USD)',
    xaxis_tickangle=-45
)

fig.show()
```
