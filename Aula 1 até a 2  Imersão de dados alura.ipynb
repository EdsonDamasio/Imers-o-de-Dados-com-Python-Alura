```python
df["senioridade"] = df["senioridade"].replace({
    'SE': 'Sênior',
    'MI': 'Pleno/Intermediário',
    'EN': 'Júnior/Iniciante',
    'EX': 'Executivo'
})

print("Categorias da coluna 'senioridade' traduzidas com sucesso!")
df['senioridade'].value_counts()
```

```python
df["employment_type"] = df["employment_type"].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

print("Categorias da coluna 'employment_type' traduzidas com sucesso!")
df['employment_type'].value_counts()
```

```python
import numpy as np
import pandas as pd

df_salarios = pd.DataFrame({
    'nome': ["Leticia", "Leandro", "Kaique", "Marcos","Everson"],
    'salario': [4000, np.nan, 5000, np.nan,7000]
})

#calcula a média salarial e substitui as nulas pela média e arrendonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(
    df_salarios['salario'].mean())

df_salarios['salario_mediana'] = df_salarios['salario'].fillna(
    df_salarios['salario'].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["segunda", "terça", "quarta", "quinta", "sexta", "sábado", "domingo"],
    "Temperatura": [30, np.nan, np.nan, 28, 27, 30, 28]
})

df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["segunda", "terça", "quarta", "quinta", "sexta", "sábado", "domingo"],
    "Temperatura": [30, np.nan, np.nan, 28, 27, 30, 28]
})

df_temperaturas["preenchido_bfill"] = df_temperaturas["Temperatura"].bfill()
df_temperaturas
```
