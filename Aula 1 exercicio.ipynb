```
#exercicio primeira imersão de dados da alura
import pandas as pd
```

```
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```
df.describe()
```

```
print('Descrição estatística do DataFrame (traduzida para português):')
describe_pt = df.describe()
describe_pt.index = [
    'contagem',
    'média',
    'desvio padrão',
    'mínimo',
    '25%',
    '50%',
    '75%',
    'máximo'
]
print(describe_pt)

# Calcular o salário médio por nível de experiência
salario_medio_por_nivel = df.groupby('experience_level')['salary_in_usd'].mean().reset_index()
salario_medio_por_nivel = salario_medio_por_nivel.rename(columns={'salary_in_usd': 'salário_médio'})

# Criar a nova coluna 'salário_médio' no DataFrame original
df = df.merge(salario_medio_por_nivel, on='experience_level', how='left')

print('\nDataFrame com a nova coluna "salário_médio":')
print(df.head())
```
